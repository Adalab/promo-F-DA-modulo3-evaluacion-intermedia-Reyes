3. Preparación de los datos:

Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas.

```python
df.columns = [column.lower() for column in df.columns]
```

```python
df.columns
```

Haciendo el análisis exploratorio te deberías haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Crea una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
def quitar_comas(cadena):
        try: 
            return float(cadena.replace(",", "."))
        except: 
            return np.nan
```

```python
columnas = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

for col in columnas: 
    df[col] = df[col].apply(quitar_comas)
```

Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada métrica ("danceability", "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.

```python
album_dance_media=df_album.groupby('artist')['danceability'].mean().reset_index()
album_dance_media
```

```python
album_dance_desviacion=df_album.groupby('artist')['danceability'].std().reset_index()
```

```python
df_dance=album_dance_media.merge(album_dance_desviacion, how='inner', on='artist', suffixes= ('media','desviacion'))

df_dance.head(3)
```

```python
album_ac_media=df_album.groupby('artist')['acousticness'].mean().reset_index()
album_ac_desviacion=df_album.groupby('artist')['acousticness'].std().reset_index()
```

```python
df_acus=album_ac_media.merge(album_ac_desviacion, how='inner', on='artist', suffixes= ('media','desviacion'))

df_acus.sample(5)
```

```python
album_speech_media=df_album.groupby('artist')['speechiness'].mean().reset_index()
album_speech_desviacion=df_album.groupby('artist')['speechiness'].std().reset_index()
```

```python
df_speech=album_speech_media.merge(album_speech_desviacion, how='inner', on='artist', suffixes= ('media','desviacion'))

df_speech.sample(5)
```
